##**Aula 05 - Machine Learning (Dados ENEN)**

```
dados_notas.head()
```

```
x_4_notas = dados_notas[['ciencias_naturais', 'ciencias_humanas', 'matematica', 'redacao']]
x_4_notas.head()
```

```
# A partir dos dados de ciencias naturais e humanas, matematica e redacao quero adivinhar as notas de linguagem e codigo 
y_adivinhar = dados_notas ['linguagem_codigo']
y_adivinhar
```

```
from sklearn.model_selection import train_test_split
import numpy as np

np.random.seed(43267)
# f(x) = y
x_treino , x_teste, y_treino, y_teste = train_test_split(x_4_notas,y_adivinhar, random_state = 326784)

print(x_treino.shape)
print(x_teste.shape)
print(y_treino.shape)
print(y_teste.shape)
```

```
x_treino.head()
```

```
from sklearn.svm import LinearSVR

modelo = LinearSVR()
modelo.fit(x_treino, y_treino)

#Parametro C é muito importante e quanto maior melhor!
```

```
#prediçoes feitas segundo o meu modelo
predicoes_notas_linguagem = modelo.predict(x_teste)
predicoes_notas_linguagem[:5]
```

```
# notas que de fato aconteceram
y_teste[:5]
```

```
import matplotlib.pyplot as plt

# Plotando a nota de fato e a predição do modelo
plt.figure(figsize=(10,10))
sns.scatterplot (x=y_teste.values, y=predicoes_notas_linguagem)
```

```
import matplotlib.pyplot as plt

# Plotando a diferença entre a nota de fato e a predição do modelo
plt.figure(figsize=(10,10))
sns.scatterplot (x=y_teste.values, y=y_teste.values - predicoes_notas_linguagem)
```

```
y_teste - x_teste.mean(axis=1)

import matplotlib.pyplot as plt

# Fazendo um modelo mais simples usando a média entre as notas simplesmente para inferir a nota da redação (Predição TOSCA)
plt.figure(figsize=(10,10))
sns.scatterplot (x=y_teste.values, y=y_teste - x_teste.mean(axis=1))
```

```


import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
sns.scatterplot (x=y_teste.values, y=y_teste - y_treino.mean())
```

```
from sklearn.tree import DecisionTreeRegressor

modelo = DecisionTreeRegressor()
modelo.fit(x_treino , y_treino)
```
